```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("8_df.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
si = SimpleImputer(strategy='mean')
```

```python
df['A'] = si.fit_transform(df[['A']])
```

```python
df.head()
```

```python
si1 = SimpleImputer(strategy='most_frequent')
df['B'] = si1.fit_transform(df[['B']])
```

```python
df.head()
```

```python
si3 = SimpleImputer(strategy='median')
df['C'] = si3.fit_transform(df[['C']])
```
